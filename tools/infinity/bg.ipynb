```python
import requests
from PIL import Image
from io import BytesIO
import os

# Local file paths
input_file = "bg.txt"
dark_output_file = "dark.txt"
light_output_file = "light.txt"

# Brightness threshold
brightness_threshold = 125  # Adjust as needed

def calculate_brightness(image):
    """Calculate the average brightness of an image."""
    grayscale = image.convert("L")
    brightness = sum(grayscale.getdata()) / len(grayscale.getdata())
    return brightness

dark_image_urls = []
light_image_urls = []

# Check if input file exists
if not os.path.exists(input_file):
    print(f"Error: File '{input_file}' not found.")
    exit()

# Read image URLs from the input file
with open(input_file, "r") as file:
    urls = file.readlines()

for url in urls:
    url = url.strip()
    if not url:
        continue
    try:
        # Download the image
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        img = Image.open(BytesIO(response.content))
        
        # Calculate brightness
        brightness = calculate_brightness(img)
        
        # Save the URL in the appropriate list
        if brightness < brightness_threshold:
            dark_image_urls.append(url)
            print(f"Dark image: {url} (Brightness: {brightness})")
        else:
            light_image_urls.append(url)
            print(f"Light image: {url} (Brightness: {brightness})")
    except Exception as e:
        print(f"Error processing {url}: {e}")

# Write dark image URLs to the dark output file
with open(dark_output_file, "w") as dark_file:
    dark_file.write("\n".join(dark_image_urls))

# Write light image URLs to the light output file
with open(light_output_file, "w") as light_file:
    light_file.write("\n".join(light_image_urls))

print(f"Saved {len(dark_image_urls)} dark image URLs to '{dark_output_file}'.")
print(f"Saved {len(light_image_urls)} light image URLs to '{light_output_file}'.")
```
