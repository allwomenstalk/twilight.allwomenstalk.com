```python
import json

# Function to read JSON data from a file
def load_json_file(filename):
    with open(filename, 'r') as file:
        return json.load(file)

# Function to add custom domain to each host entry
def add_custom_domain(hosts):
    for host in hosts:
        # Extract the subdomain part from the name
        subdomain_part = host["name"].split("-")[-1]
        new_domain = f"{subdomain_part}.allwomenstalk.com"
        host['domain'] = new_domain
        # Add this new domain if it is not already in the domains list
        if new_domain not in host["domains"]:
            host["domains"].append(new_domain)

    return hosts

# Load the initial data from hosts.json
data = load_json_file('hosts.json')

# Update the JSON data
updated_data = add_custom_domain(data)

# Convert the updated data back to JSON string
updated_json_string = json.dumps(updated_data, indent=4)
print(updated_json_string)

# Optionally, write the updated JSON back to the file
with open('hosts.json', 'w') as file:
    json.dump(updated_data, file, indent=4)
```
